```python
sales = pd.read_csv('kc_house_data.csv')
sales = sales.sort_values(by=['sqft_living','price'])
```

```python
set1 = pd.read_csv('wk3_kc_house_set_1_data.csv')
set2 = pd.read_csv('wk3_kc_house_set_2_data.csv')
set3 = pd.read_csv('wk3_kc_house_set_3_data.csv')
set4 = pd.read_csv('wk3_kc_house_set_4_data.csv')
validData = pd.read_csv('wk3_kc_house_valid_data.csv')
testData = pd.read_csv('wk3_kc_house_test_data.csv')
trainData = pd.read_csv('wk3_kc_house_train_data.csv')
validData = validData.sort_values(by=['sqft_living','price'])
trainData = trainData.sort_values(by=['sqft_living','price'])
testData = testData.sort_values(by=['sqft_living','price'])
```

```python
def polyFeat(feat, degree):
    df = pd.DataFrame()
    if degree > 0:
        for pwr in range(1, degree+1):
            name = 'power_' + str(pwr)
            df[name] = feat ** pwr
    return df
```

```python
poly1Data = polyFeat(sales['sqft_living'], 1)
poly1Data['price'] = sales.price
```

```python
poly1Data
```

```python
model1 = LinearRegression()
model1.fit(poly1Data[['power_1']], poly1Data['price'])
```

```python
plt.plot(poly1Data['power_1'], poly1Data['price'], '.', poly1Data['power_1'], model1.predict(poly1Data[['power_1']]), '-')
```

```python
sets = [set1, set2, set3, set4]
poly15data = []
poly15Feats = []
models15 = []
for i in range(4):
    poly15data.append(polyFeat(sets[i]['sqft_living'], 15))
    poly15Feats.append(poly15data[i].columns)
    poly15data[i]['price'] = sets[i]['price']
    models15.append(LinearRegression())
    models15[i].fit(poly15data[i][poly15Feats[i]], poly15data[i]['price'])
    print('for set ', str(i+1), ' coef = ' ,str(models15[i].coef_[14]))
    plt.figure()
    plt.plot(poly15data[i]['power_1'], poly15data[i]['price'], '.', poly15data[i]['power_1'], models15[i].predict(poly15data[i][poly15Feats[i]]), '-')
```

```python
def getRSS(output, predictions):
    RSS = np.sum(np.square(output-predictions))
#     RSS = RSS ** 0.5
    return RSS
```

```python
polyDatas = [0]
polyDatasValid = [0]
polyFeats = [0]
models = [0]
predcitions = [0]
RSS = [0]
for pwr in range(1,16):
    polyDatas.append(polyFeat(trainData['sqft_living'], pwr))
    polyDatasValid.append(polyFeat(validData['sqft_living'], pwr))
    polyFeats.append(polyDatas[pwr].columns)
    models.append(LinearRegression())
    models[pwr].fit(polyDatas[pwr], trainData['price'])
    predcitions.append(models[pwr].predict(polyDatasValid[pwr]))
    RSS.append(getRSS(validData['price'], predcitions[pwr]))
```

```python
for i in range(1, 16):
    print('%d : %f' % (i, RSS[i]))
```

```python
polyDataTest = polyFeat(testData.sqft_living, 7)
predictions = models[7].predict(polyDataTest)
RSSTest = getRSS(testData.price,predictions)
```
