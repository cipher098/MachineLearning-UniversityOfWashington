```python
def polyFeat(feat, degree):
    df = pd.DataFrame()
    if degree > 0:
        for pwr in range(1, degree+1):
            name = 'power_' + str(pwr)
            df[name] = feat ** pwr
    return df
```

```python
poly15 = polyFeat(sales.sqft_living, 15)
```

```python
l2pen = 1.5e-5
```

```python
model = Ridge(alpha=l2pen, normalize=True)
model.fit(poly15, sales.price)
```

```python
sets = [set1, set2, set3, set4]
poly15data = []
poly15Feats = []
models15 = []
for i in range(4):
    poly15data.append(polyFeat(sets[i]['sqft_living'], 15))
    poly15Feats.append(poly15data[i].columns)
    models15.append(Ridge(alpha=l2pen1, normalize=True))
    models15[i].fit(poly15data[i], sets[i]['price'])
    print('for set ', str(i+1), ' coef = ' ,str(models15[i].coef_[0]))
    plt.figure()
    plt.plot(poly15data[i]['power_1'], sets[i]['price'], '.', poly15data[i]['power_1'], models15[i].predict(poly15data[i]), '-')
```

```python
def getRSS(output, predictions):
    RSS = np.sum(np.square(output-predictions))
#     RSS = RSS ** 0.5
    return RSS
```

```python
def compValErr(k, l2p, df, op):
    n = len(df)
    RSSValid = 0
    empty_vector = np.empty(k)
    for i in range(k):
        start = floor((n*i)/k)
        end = floor((n*(i+1))/k-1)
        print(start, end)
        validSet = df[start:end+1]
        validOp = op[start:end+1]
        trainSet = df[end+1:n]
        trainSet = trainSet.append(df[0:start])
        trainOp = op[end+1:n]
        trainOp = trainOp.append(op[0:start])
        feats = trainSet.columns
        trainSet['price'] = trainOp
#         model = Ridge(alpha=l2p, normalize=True)
        model = graphlab.linear_regression.create(trainSet, target = 'price', features = feats,l2_penalty=l2p, validation_set = None)
        
        print('len ',len(trainSet), len(trainOp))
        prediction = model.predict(validSet)
        empty_vector[i] =  getRSS(validOp, prediction)
    return empty_vector.mean()   
```

```python
poly15Tr = polyFeat(trainValid['sqft_living'], 15)
valErr = {}
k = 10
for l2p in np.logspace(1, 9, num=17):
    valErr[str(l2p)] = compValErr(k, l2p, poly15Tr, trainValid.price)
```
